```python
# Cell 3: Enhanced Study Assistant Class with Colorful UI
class EnhancedStudyAssistant:
    def __init__(self, model):
        self.model = model
        print_success("Enhanced Study Assistant initialized! 🎓")

    def extract_text_from_pdf(self, pdf_file):
        """Extract text from uploaded PDF file"""
        try:
            reader = PyPDF2.PdfReader(pdf_file)
            text = ""
            total_pages = len(reader.pages)

            print_info(f"📄 Processing PDF with {total_pages} pages...")

            for i, page in enumerate(reader.pages, 1):
                text += page.extract_text() + "\n"
                if i % 5 == 0:  # Progress update every 5 pages
                    print_info(f"   Processed {i}/{total_pages} pages...")

            print_success(f"PDF processed successfully! Extracted {len(text)} characters")
            return text
        except Exception as e:
            print_error(f"Error reading PDF: {e}")
            return None

    def extract_text_from_docx(self, docx_file):
        """Extract text from uploaded Word document"""
        try:
            print_info("📝 Processing Word document...")

            # Method 1: Using docx2txt
            text = docx2txt.process(docx_file)
            if text:
                print_success(f"Word document processed! Extracted {len(text)} characters")
                return text

            # Method 2: Using python-docx as fallback
            docx_file.seek(0)
            doc = Document(docx_file)
            text = ""

            print_info("   Extracting paragraphs...")
            for paragraph in doc.paragraphs:
                text += paragraph.text + "\n"

            print_info("   Extracting tables...")
            for table in doc.tables:
                for row in table.rows:
                    for cell in row.cells:
                        text += cell.text + " "
                    text += "\n"

            print_success(f"Word document processed! Extracted {len(text)} characters")
            return text
        except Exception as e:
            print_error(f"Error reading Word document: {e}")
            return None

    def detect_file_type(self, filename):
        """Detect file type based on extension"""
        filename = filename.lower()
        if filename.endswith('.pdf'):
            return 'pdf'
        elif filename.endswith(('.docx', '.doc')):
            return 'docx'
        else:
            return 'unknown'

    def extract_text_from_file(self, file_content, filename):
        """Extract text from any supported file format"""
        print_header(f"📁 PROCESSING FILE: {filename}")

        file_type = self.detect_file_type(filename)
        file_obj = io.BytesIO(file_content)

        if file_type == 'pdf':
            return self.extract_text_from_pdf(file_obj)
        elif file_type == 'docx':
            return self.extract_text_from_docx(file_obj)
        else:
            print_error(f"Unsupported file type: {filename}")
            print_info("📋 Supported formats: PDF (.pdf), Word (.docx, .doc)")
            return None

    def summarize_content(self, content: str) -> str:
        """Summarize the study material into key points"""
        print_header("📋 GENERATING SUMMARY")
        print_info("🤖 AI is analyzing your content...")

        prompt = f"""
Please summarize the following study material into concise bullet points that capture the key concepts and important information:

Study Material:
{content}

Instructions:
- Create 5-8 clear, concise bullet points
- Focus on the most important concepts
- Each point should be 1-2 sentences maximum
- Use simple, clear language

Summary:
"""

        try:
            response = self.model.generate_content(prompt)
            print_success("Summary generated successfully! ✨")
            return response.text
        except Exception as e:
            print_error(f"Error generating summary: {e}")
            return "Unable to generate summary"

    def generate_quiz_questions(self, content: str, num_questions: int = 5) -> List[Dict]:
        """Generate multiple-choice quiz questions based on the content"""
        print_header(f"🧠 GENERATING {num_questions} QUIZ QUESTIONS")
        print_info("🤖 AI is creating your personalized quiz...")

        prompt = f"""
Based on the following study material, create {num_questions} multiple-choice quiz questions.

Study Material:
{content}

Instructions:
- Create {num_questions} multiple-choice questions
- Each question should have 4 options (a, b, c, d)
- Questions should test understanding of key concepts
- Include the correct answer for each question
- Vary the difficulty from basic recall to application
- Format each question clearly

Please format your response as follows for each question:
Question X: [Question text]
a) [Option A]
b) [Option B]
c) [Option C]
d) [Option D]
Correct Answer: [Letter]
Explanation: [Brief explanation]

Quiz Questions:
"""

        try:
            response = self.model.generate_content(prompt)
            questions = self._parse_quiz_response(response.text)

            if questions:
                print_success(f"Quiz generated successfully! Created {len(questions)} questions ✨")
            else:
                print_warning("Quiz generation completed but parsing failed")

            return questions
        except Exception as e:
            print_error(f"Error generating quiz: {e}")
            return []

    def _parse_quiz_response(self, response_text: str) -> List[Dict]:
        """Parse the quiz response into structured format"""
        print_info("🔄 Parsing quiz questions...")

        questions = []
        question_blocks = re.split(r'Question \d+:', response_text)[1:]

        for i, block in enumerate(question_blocks, 1):
            try:
                lines = block.strip().split('\n')
                question_text = lines[0].strip()

                options = {}
                correct_answer = ""
                explanation = ""

                for line in lines[1:]:
                    line = line.strip()
                    if line.startswith(('a)', 'b)', 'c)', 'd)')):
                        key = line[0]
                        value = line[3:].strip()
                        options[key] = value
                    elif line.startswith('Correct Answer:'):
                        correct_answer = line.split(':')[1].strip().lower()
                    elif line.startswith('Explanation:'):
                        explanation = line.split(':', 1)[1].strip()

                if question_text and options and correct_answer:
                    questions.append({
                        'question': question_text,
                        'options': options,
                        'correct_answer': correct_answer,
                        'explanation': explanation
                    })
                    print_info(f"   ✓ Question {len(questions)} parsed successfully")
            except Exception as e:
                print_warning(f"Error parsing question {i}: {e}")
                continue

        return questions

    def display_quiz(self, questions: List[Dict]):
        """Display quiz questions in a colorful, user-friendly format"""
        print_header("🎯 YOUR PERSONALIZED QUIZ")

        for i, q in enumerate(questions, 1):
            # Question header with color
            print_colored(f"\n🔹 Question {i}: {q['question']}", "96")  # Cyan
            print()

            # Options with different colors
            colors = ["93", "94", "95", "92"]  # Yellow, Blue, Magenta, Green
            for idx, (key, value) in enumerate(q['options'].items()):
                print_colored(f"   {key.upper()}) {value}", colors[idx % 4])

            # Correct answer in bright green
            print_colored(f"\n✅ Correct Answer: {q['correct_answer'].upper()}", "92")

            # Explanation in blue
            if q['explanation']:
                print_colored(f"💡 Explanation: {q['explanation']}", "94")

            print_colored("-" * 60, "90")  # Gray separator

# Initialize the Enhanced Study Assistant
study_assistant = EnhancedStudyAssistant(model)
print_success("Enhanced Study Assistant ready to rock! 🚀")
```

```python
# Cell 4: Enhanced File Processing with Progress Indicators
def upload_and_process_file():
    """Upload and process PDF or Word document with enhanced UI"""
    print_header("📁 FILE UPLOAD CENTER")
    print_info("📋 Supported formats: PDF (.pdf), Word (.docx, .doc)")
    print_colored("🎨 Drop your file below! ⬇️", "95")

    uploaded = files.upload()

    if not uploaded:
        print_error("No file uploaded")
        return None, None

    filename = list(uploaded.keys())[0]
    file_content = uploaded[filename]

    print_success(f"File received: {filename}")
    print_info(f"📊 File size: {len(file_content):,} bytes ({len(file_content)/1024:.1f} KB)")

    # Progress bar simulation
    import time
    print_info("🔄 Processing file...")
    for i in range(0, 101, 20):
        print(f"\rProgress: [{'█' * (i//5)}{'░' * (20-i//5)}] {i}%", end='')
        time.sleep(0.1)
    print()  # New line

    # Extract text
    text = study_assistant.extract_text_from_file(file_content, filename)

    if text:
        text = text.strip()
        word_count = len(text.split())

        print_success(f"Text extraction completed!")
        print_info(f"📈 Statistics: {len(text):,} characters, {word_count:,} words")

        if len(text) < 50:
            print_warning("Extracted text is very short. Please verify file content.")

        # Preview with color
        preview = text[:300] + "..." if len(text) > 300 else text
        print_colored("\n📖 CONTENT PREVIEW:", "96")
        print_colored("─" * 40, "90")
        print_colored(preview, "97")  # White text
        print_colored("─" * 40, "90")

        return text, filename
    else:
        print_error("Failed to extract text from file")
        print_warning("🔧 Troubleshooting tips:")
        print_info("   • Ensure file is not password protected")
        print_info("   • Check if file contains readable text")
        print_info("   • Try a different file format")
        return None, None

def process_text_input():
    """Process text input with enhanced interface"""
    print_header("✍️ TEXT INPUT CENTER")
    print_info("📝 Enter your study material below")
    print_colored("💡 Tip: Paste content from any source!", "93")
    print_colored("Type 'END' on a new line when finished", "94")
    print_colored("─" * 50, "90")

    lines = []
    line_count = 0

    while True:
        try:
            line = input(f"{line_count + 1:3d}│ ")
            if line.strip().upper() == 'END':
                break
            lines.append(line)
            line_count += 1

            # Progress indicator every 10 lines
            if line_count % 10 == 0:
                print_info(f"   📊 {line_count} lines entered so far...")

        except KeyboardInterrupt:
            print_error("\nInput cancelled")
            return None

    text = '\n'.join(lines).strip()
    if text:
        word_count = len(text.split())
        print_success("Text input completed!")
        print_info(f"📈 Statistics: {len(text):,} characters, {word_count:,} words")
        return text
    else:
        print_error("No text provided")
        return None

print_success("Enhanced file processing functions loaded! 📁✨")
```

```python
def upload_and_process_file():
    """Upload and process PDF or Word document"""
    print("📁 Please upload your study material:")
    print("📋 Supported formats: PDF (.pdf), Word (.docx, .doc)")

    uploaded = files.upload()

    if not uploaded:
        print("❌ No file uploaded")
        return None, None

    filename = list(uploaded.keys())[0]
    file_content = uploaded[filename]

    print(f"✅ File uploaded: {filename}")
    print(f"📊 File size: {len(file_content)} bytes")

    # Detect and process the file
    text = study_assistant.extract_text_from_file(file_content, filename)

    if text:
        text = text.strip()
        if len(text) < 50:
            print("⚠️  Warning: Extracted text is very short. Please check if the file contains readable text.")

        word_count = len(text.split())
        print(f"✅ Successfully extracted {len(text)} characters ({word_count} words)")

        # Show preview of extracted text
        preview = text[:200] + "..." if len(text) > 200 else text
        print(f"\n📖 Text preview:\n{preview}")

        return text, filename
    else:
        print("❌ Failed to extract text from file")
        print("🔧 Troubleshooting tips:")
        print("   - Ensure the file is not password protected")
        print("   - Check if the file contains readable text (not just images)")
        print("   - Try a different file format")
        return None, None

def process_text_input():
    """Process text input directly"""
    print("📝 Enter your study material (paste text below):")
    print("💡 Tip: You can paste content from any source")
    print("Type 'END' on a new line when finished:")
    print("-" * 40)

    lines = []
    while True:
        try:
            line = input()
            if line.strip().upper() == 'END':
                break
            lines.append(line)
        except KeyboardInterrupt:
            print("\n❌ Input cancelled")
            return None

    text = '\n'.join(lines).strip()
    if text:
        word_count = len(text.split())
        print(f"✅ Successfully captured {len(text)} characters ({word_count} words)")
        return text
    else:
        print("❌ No text provided")
        return None

def validate_content(text):
    """Validate if the content is suitable for quiz generation"""
    if not text or len(text.strip()) < 50:
        print("⚠️  Content is too short for effective quiz generation.")
        print("💡 Please provide at least a few sentences of study material.")
        return False

    word_count = len(text.split())
    if word_count < 20:
        print("⚠️  Content has very few words. Quiz quality may be limited.")
        return False

    return True

print("✅ Enhanced upload and processing functions ready!")
print("📋 Supported formats: PDF, Word documents, and direct text input")
```

```python
# Cell 5: Enhanced Main Application with Colorful Interface
def run_enhanced_study_assistant():
    """Main function with beautiful colorful interface"""

    # Welcome banner
    print_colored("\n" + "█" * 70, "95")
    print_colored("█" + " " * 68 + "█", "95")
    print_colored("█" + "🎓 WELCOME TO ENHANCED STUDY ASSISTANT QUIZ GENERATOR! 🎓".center(68) + "█", "95")
    print_colored("█" + " " * 68 + "█", "95")
    print_colored("█" * 70, "95")

    print_colored("\n🌟 Transform your study materials into interactive quizzes! 🌟", "96")
    print_colored("🤖 Powered by Google Gemini AI ✨", "94")

    # Method selection with colors
    print_header("📋 CHOOSE YOUR INPUT METHOD")
    print_colored("1. 📤 Upload PDF/Word file", "92")
    print_colored("2. ✏️  Enter text directly", "93")

    while True:
        choice = input(print_colored("\n🎯 Enter your choice (1 or 2): ", "96") or "").strip()
        if choice in ['1', '2']:
            break
        print_warning("Please enter 1 or 2")

    study_material = None
    filename = None

    if choice == '1':
        study_material, filename = upload_and_process_file()
    elif choice == '2':
        study_material = process_text_input()

    if not study_material:
        print_error("No study material to process. Exiting...")
        return

    # Content validation with progress
    print_header("🔍 CONTENT VALIDATION")
    print_info("Checking content quality...")

    word_count = len(study_material.split())
    char_count = len(study_material)

    if char_count < 100:
        print_warning("Content is quite short. Quiz quality may be limited.")
    elif char_count < 500:
        print_info("Content length is adequate for quiz generation.")
    else:
        print_success("Excellent content length for high-quality quiz generation!")

    print_info(f"📊 Final stats: {char_count:,} characters, {word_count:,} words")

    # Generate summary
    print("\n" + "🎨" * 30)
    summary = study_assistant.summarize_content(study_material)

    print_header("📋 YOUR STUDY SUMMARY")
    print_colored(summary, "97")

    # Quiz configuration
    print_header("⚙️ QUIZ CONFIGURATION")

    while True:
        try:
            num_questions = int(input(print_colored("🤔 How many quiz questions? (1-10): ", "96") or "5"))
            if 1 <= num_questions <= 10:
                break
            else:
                print_warning("Please enter a number between 1 and 10")
        except ValueError:
            print_warning("Please enter a valid number")

    print_success(f"Perfect! Generating {num_questions} questions for you...")

    # Generate quiz with progress
    print_info("🤖 AI is crafting your personalized quiz...")
    import time

    # Simulate AI thinking with animated dots
    for i in range(3):
        for dots in ["⠋", "⠙", "⠹", "⠸", "⠼", "⠴", "⠦", "⠧", "⠇", "⠏"]:
            print(f"\r🧠 Generating questions {dots}", end="", flush=True)
            time.sleep(0.1)
    print("\r" + " " * 30 + "\r", end="")  # Clear line

    questions = study_assistant.generate_quiz_questions(study_material, num_questions)

    if questions:
        study_assistant.display_quiz(questions)

        # Save option with enhanced UI
        print_header("💾 SAVE YOUR QUIZ")
        print_colored("Would you like to download your quiz results?", "96")
        print_colored("1. 📥 Yes, download my quiz!", "92")
        print_colored("2. 🚫 No thanks, I'm done", "93")

        save_choice = input(print_colored("\nYour choice (1 or 2): ", "96") or "").strip()

        if save_choice == '1':
            save_enhanced_quiz(summary, questions, filename)

        print_success("🎉 Quiz generation completed successfully!")
        print_colored("✨ Hope you ace your studies! Good luck! 🍀", "95")

    else:
        print_error("Failed to generate quiz questions")
        print_info("💡 Try with different content or check your internet connection")

def save_enhanced_quiz(summary, questions, original_filename=None):
    """Save quiz with enhanced formatting and progress indicator"""
    print_info("📝 Preparing your quiz file...")

    # Progress simulation
    import time
    for i in range(0, 101, 25):
        print(f"\rSaving progress: [{'█' * (i//4)}{'░' * (25-i//4)}] {i}%", end='')
        time.sleep(0.2)
    print()

    timestamp = time.strftime("%Y%m%d_%H%M%S")
    filename = f"enhanced_quiz_{timestamp}.txt"

    content = f"""
{'=' * 70}
🎓 ENHANCED STUDY ASSISTANT QUIZ RESULTS 🎓
{'=' * 70}
Generated on: {time.strftime("%Y-%m-%d %H:%M:%S")}
Original file: {original_filename or 'Text input'}
{'=' * 70}

📋 STUDY MATERIAL SUMMARY
{'-' * 30}
{summary}

🧠 QUIZ QUESTIONS
{'-' * 30}
"""

    for i, question in enumerate(questions, 1):
        content += f"""
🔹 Question {i}: {question['question']}

Options:
"""
        for key, value in question['options'].items():
            marker = "✅" if key == question['correct_answer'] else "  "
            content += f"   {marker} {key.upper()}) {value}\n"

        content += f"""
✅ Correct Answer: {question['correct_answer'].upper()}
💡 Explanation: {question['explanation']}
{'-' * 50}
"""

    content += f"""
{'=' * 70}
🌟 Generated by Enhanced Study Assistant
🤖 Powered by Google Gemini AI
✨ Happy studying!
{'=' * 70}
"""

    # Save file
    with open(filename, 'w', encoding='utf-8') as f:
        f.write(content)

    print_success(f"Quiz saved as: {filename}")

    # Download file
    files.download(filename)
    print_success("📥 File downloaded successfully!")
    print_info("🎯 You can now access your quiz anytime!")

print_success("Enhanced main application loaded! 🚀")
```
